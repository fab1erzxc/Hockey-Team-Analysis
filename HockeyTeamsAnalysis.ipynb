```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hockey.csv')
df
```

```python
df.head(10)
```

```python
teams = df['Team Name'].unique()
teams
```

```python
teams_to_plot = teams

# Plot the trend of win percentages over time for selected teams
plt.figure(figsize=(10, 6))
for team in teams_to_plot:
    team_data = df[df['Team Name'] == team]
    plt.plot(team_data['Year'], team_data['Win %'], label=team)

plt.title('Win Percentage Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Win Percentage')
plt.legend()
plt.grid(True)
plt.show()
```

## Team Performance Comparison

```python
# Group the DataFrame by 'Team Name' and calculate mean statistics
team_performance = df.groupby('Team Name').agg({
    'Win %': 'mean',
    'Goals For (GF)': 'mean',
    'Goals Against (GA)': 'mean',
    '+ / -': 'mean'
}).reset_index()

# Display the summary statistics for team performance
print(team_performance)
```

```python
# Set the figure size
plt.figure(figsize=(18, 12))

# Plot mean win percentage
plt.subplot(221)
plt.bar(team_performance['Team Name'], team_performance['Win %'])
plt.title('Mean Win Percentage')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(18, 12))

# Plot mean goals scored
plt.subplot(222)
plt.bar(team_performance['Team Name'], team_performance['Goals For (GF)'])
plt.title('Mean Goals For')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(18, 12))

# Plot mean goals conceded
plt.subplot(223)
plt.bar(team_performance['Team Name'], team_performance['Goals Against (GA)'])
plt.title('Mean Goals Against')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(18, 12))

# Plot mean goal differential
plt.subplot(224)
plt.bar(team_performance['Team Name'], team_performance['+ / -'])
plt.title('Mean Goal Differential')
plt.xticks(rotation=90)
```

## Season Performance:

```python
# Group the DataFrame by 'Year' and calculate mean statistics
season_performance = df.groupby('Year').agg({
    'Win %': 'mean',
    'Goals For (GF)': 'mean',
    'Goals Against (GA)': 'mean',
    '+ / -': 'mean'
}).reset_index()

# Display the summary statistics for season performance
print(season_performance)
```

```python
# Set the figure size
plt.figure(figsize=(12, 8))

# Plot mean win percentage
plt.subplot(221)
plt.plot(season_performance['Year'], season_performance['Win %'], marker='o')
plt.title('Mean Win Percentage Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Win Percentage')

# Plot mean goals scored
plt.subplot(222)
plt.plot(season_performance['Year'],
         season_performance['Goals For (GF)'], marker='o')
plt.title('Mean Goals For Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Goals For')

# Plot mean goals conceded
plt.subplot(223)
plt.plot(season_performance['Year'],
         season_performance['Goals Against (GA)'], marker='o')
plt.title('Mean Goals Against Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Goals Against')

# Plot mean goal differential
plt.subplot(224)
plt.plot(season_performance['Year'], season_performance['+ / -'], marker='o')
plt.title('Mean Goal Differential Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Goal Differential')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

## Impact of Overtime Losses:

```python
# Check for teams with overtime losses
teams_with_ot_losses = df[df['OT Losses'].notnull()]

# Check for teams without overtime losses
teams_without_ot_losses = df[df['OT Losses'].isnull()]

# Calculate average win percentage for teams with and without overtime losses
avg_win_percentage_with_ot_losses = teams_with_ot_losses['Win %'].mean()
avg_win_percentage_without_ot_losses = teams_without_ot_losses['Win %'].mean()

# Print the average win percentage for each group
print("Average Win Percentage for Teams with Overtime Losses:",
      avg_win_percentage_with_ot_losses)
print("Average Win Percentage for Teams without Overtime Losses:",
      avg_win_percentage_without_ot_losses)
```

```python
# Set up the bar plot
plt.figure(figsize=(8, 6))

# Plot the average win percentage for teams with and without overtime losses
plt.bar(['With Overtime Losses', 'Without Overtime Losses'], [
        avg_win_percentage_with_ot_losses, avg_win_percentage_without_ot_losses], color=['red', 'blue'])
plt.title('Impact of Overtime Losses on Win Percentage')
plt.xlabel('Overtime Losses')
plt.ylabel('Average Win Percentage')

# Show plot
plt.show()
```

## Goal Differential:



```python
# Calculate goal differential (Goals For - Goals Against) for each team in each year
df['Goal Differential'] = df['Goals For (GF)'] - df['Goals Against (GA)']

# Display the first few rows of the DataFrame with the goal differential column
print(df.head())
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Plot histogram for goal differential
plt.hist(df['Goal Differential'], bins=15, color='green', alpha=0.7)
plt.title('Distribution of Goal Differential')
plt.xlabel('Goal Differential')
plt.ylabel('Frequency')
plt.grid(True)

# Show plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Plot scatter plot for goal differential vs. win percentage
plt.scatter(df['Goal Differential'], df['Win %'], color='blue', alpha=0.5)
plt.title('Goal Differential vs. Win Percentage')
plt.xlabel('Goal Differential')
plt.ylabel('Win Percentage')
plt.grid(True)

# Show plot
plt.show()
```

## Comparative Analysis:



```python
# Group the DataFrame by 'Year' and calculate mean win percentage for each year
yearly_performance = df.groupby('Year')['Win %'].mean().reset_index()

# Display the summary statistics for yearly performance
print(yearly_performance)
```

```python
# Set up the line plot
plt.figure(figsize=(10, 6))

# Plot the mean win percentage for each year
plt.plot(yearly_performance['Year'],
         yearly_performance['Win %'], marker='o', color='blue')
plt.title('Mean Win Percentage Over Years')
plt.xlabel('Year')
plt.ylabel('Mean Win Percentage')
plt.grid(True)

# Show plot
plt.show()
```

## Historical Performance:



```python
# Select key teams for historical performance analysis
key_teams = ['Boston Bruins', 'Chicago Blackhawks', 'Montreal Canadiens']

# Filter the DataFrame to include data only for the selected key teams
key_teams_df = df[df['Team Name'].isin(key_teams)]

# Group the filtered DataFrame by 'Year' and calculate mean win percentage for each year
historical_performance = key_teams_df.groupby(['Team Name', 'Year'])[
    'Win %'].mean().unstack()

# Display the historical performance DataFrame
print(historical_performance)
```

```python
# Set up the line plot
plt.figure(figsize=(10, 6))

# Plot the historical performance for each team
for team in key_teams:
    plt.plot(historical_performance.loc[team], marker='o', label=team)

plt.title('Historical Performance of Key Teams')
plt.xlabel('Year')
plt.ylabel('Mean Win Percentage')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
