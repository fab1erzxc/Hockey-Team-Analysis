```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = r'https://www.scrapethissite.com/pages/forms/?page_num=1&per_page=100'
pages =[i for i in range(1, 7)]
contents = []
for page in pages:
    url = f'https://www.scrapethissite.com/pages/forms/?page_num={page}&per_page=100'
    r = requests.get(url)
    soup = bs(r.content, 'html.parser')
    contents.append(soup)

print(contents)
```

```python
headers = []
for content in contents:
    for header in content.find_all('th'):
        headers.append(header.text.strip())

headers = headers[:9]
print(headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=headers)
df
```

```python
rows = []
for content in contents:
    for row in content.find_all('tr'):
        rows.append([i.text.strip() for i in row.find_all('td')])

rows = rows[1:]
rows[95:105]
```

```python
rows = [i for i in rows if len(i) == 9]
rows[95:105]
```
